```python
# Combine Multiple sheets into one Workbook
import pandas as pd

excel_file_path = "/content/drive/MyDrive/Colab Notebooks/Sample_Workbook.xlsx"
output_master_path = "/content/drive/MyDrive/Colab Notebooks/Master_Workbook.xlsx"

# Read all sheets into a dictionary of DataFrames
all_sheets = pd.read_excel(excel_file_path, sheet_name=None)

# Create an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Loop through each sheet and append the data to the combined_data DataFrame
for sheet_name, sheet_data in all_sheets.items():
    combined_data = combined_data.append(sheet_data, ignore_index=True)

# Write the combined data to a new Excel workbook
combined_data.to_excel(output_master_path, index=False)

print(f"All sheets combined and saved to {output_master_path}")
```
